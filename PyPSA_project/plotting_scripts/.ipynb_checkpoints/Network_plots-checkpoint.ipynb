```python
from IPython import get_ipython
#from IPython.display import display, clear_output
import pandas as pd
import numpy as np
import time
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
from scipy.spatial import ConvexHull,  Delaunay
from scipy.interpolate import griddata
import pypsa
```

```python
network_path = 'C:/Users/Tim/OneDrive - Aarhus universitet/Speciale/PyPSA_project/data/networks/euro_00'
network = pypsa.Network()
network.import_from_hdf5(network_path)
```

```python
network.generators_t.p_max_pu
```

```python
solar_pu = []
onwind_pu = []
offwind_pu = []
offwind_indexes = []
for node in network.buses.index:
    try :
        offwind_pu.append(np.mean(network.generators_t.p_max_pu[node+' offwind']))
        offwind_indexes.append(node)
    except :
        pass

    try :
        onwind_pu.append(np.mean(network.generators_t.p_max_pu[node+' onwind']))
    except : 
        onwind_pu.append(0)
    try :
        solar_pu.append(np.mean(network.generators_t.p_max_pu[node+' solar']))
    except : 
        solar_pu.append(0)
        
solar_pu = np.array(solar_pu)
offwind_pu = np.array(offwind_pu)
onwind_pu = np.array(onwind_pu)
```

```python
from iso3166 import countries

fig = make_subplots(
    rows=1, cols=3,
    #subplot_titles = ['Offshore wind','Onshore wind','Solar'],
    specs=[[{"type": 'Choropleth'}, {"type": 'Choropleth'}, {'type':'Choropleth'}],])

country_area = pd.read_csv('C:/Users\Tim\OneDrive - Aarhus universitet\Speciale\PyPSA_project\data/country_sizes.csv')

#codes = ['AUT', 'BEL', 'BGR', 'BIH', 'HRV', 'CHE', 'CZE', 'DNK', 'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LTU', 'LUX', 'NLD', 'NOR', 'POL', 'PRT', 'ROU', 'SRB','SVK', 'SVN', 'ESP', 'SWE', 'GBR']
codes = [countries.get(alpha2).alpha3 for alpha2 in network.buses.index]
offwind_codes = [countries.get(alpha2).alpha3 for alpha2 in offwind_indexes]
names = [countries.get(alpha2).name for alpha2 in network.buses.index]
#values = network.generators.p_nom_max[network.generators.carrier == 'onwind']
area = [(country_area[country_area['Country Code']==country]['2018']).values[0] for country in codes]

fig.add_trace(go.Choropleth(
                    geo='geo1',
                    locations = offwind_codes,
                    z = offwind_pu,#/area,
                    text = names,
                    colorscale = 'Thermal',
                    autocolorscale=False,
                    zmax=0.5,
                    zmin=0,
                    reversescale=False,
                    marker_line_color='darkgray',
                    marker_line_width=0.5,
                    #colorbar_tickprefix = '',
                    #colorbar_title = 'Potential [MWh/km^2]'
                        ),row=1,col=1) 

fig.add_trace(go.Choropleth(
                    geo='geo2',
                    locations = codes,
                    z = onwind_pu,#/area,
                    text = names,
                    colorscale = 'Thermal',
                    autocolorscale=False,
                    zmax=0.5,
                    zmin=0,
                    reversescale=False,
                    marker_line_color='darkgray',
                    marker_line_width=0.5,
                    #colorbar_tickprefix = '',
                    #colorbar_title = 'Potential [MWh/km^2]'
                        ),row=1,col=2)
fig.add_trace(go.Choropleth(
                    geo='geo3',
                    locations = codes,
                    z = solar_pu,#/area,
                    text = names,
                    colorscale = 'Thermal',
                    autocolorscale=False,
                    zmax=0.5,
                    zmin=0,
                    reversescale=False,
                    marker_line_color='darkgray',
                    marker_line_width=0.5,
                    #colorbar_tickprefix = '',
                    #colorbar_title = 'Average availability'
                        ),row=1,col=3)


fig.update_geos(
        scope = 'europe',
        projection_type = 'azimuthal equal area',
        showland = True,
        landcolor = 'rgb(243, 243, 243)',
        countrycolor = 'rgb(204, 204, 204)',
        lataxis = dict(
            range = [35, 64],
            showgrid = False
        ),
        lonaxis = dict(
            range = [-11, 26],
            showgrid = False
        )
    ),



fig.update_layout(
    autosize=False,
    width=900,
    height=500,
    showlegend=False,
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)
fig.update_layout(
    geo1=dict(
        domain=dict(x=[0,0.33],y=[0,1])),#Top Rigth
    geo2=dict(
        domain=dict(x=[0.33,0.66],y=[0,1])),#Top Left
    geo3=dict(
        domain=dict(x=[0.66,1],y=[0,1])),#Botom left
    annotations= [dict(
    xref= 'paper',
    yref= 'paper',
    x= 0.2,
    xanchor= 'right',
    y= 0,
    yanchor= 'bottom',
    text= 'Offshore wind',
    showarrow= False
    ),
    dict(
    xref= 'paper',
    yref= 'paper',
    x= 0.57,
    xanchor= 'right',
    y= 0,
    yanchor= 'bottom',
    text= 'Onshore wind',
    showarrow= False
    ),
     dict(
    xref= 'paper',
    yref= 'paper',
    x= 0.9,
    xanchor= 'right',
    y= 0,
    yanchor= 'bottom',
    text= 'Solar',
    showarrow= False
    ),
    dict(
    xref= 'paper',
    yref= 'paper',
    x= 0.6,
    xanchor= 'right',
    y= 0.9,
    yanchor= 'bottom',
    text= 'Average capacity factor',
    showarrow= False
    ),]
)

fig.show()
fig.write_image("C:/Users\Tim\OneDrive - Aarhus universitet\Speciale\Report\Images\wind_availability.pdf")
```
