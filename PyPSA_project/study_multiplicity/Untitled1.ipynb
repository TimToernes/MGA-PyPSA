```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

class PolynomialRegression:
    def __init__(self,deg=3):
        self.poly = PolynomialFeatures(degree=deg,include_bias=False)
        self.lin_reg = LinearRegression()
    def fit(self,x,y):
        self.poly_variables= self.poly.fit_transform(x)
        self.lin_reg.fit(self.poly_variables,y)
        return 
    def transform(self,x):
        x_poly_variables = self.poly.transform(x)
        y = self.lin_reg.predict(x_poly_variables)
        return y
    def coef(self):
        print(self.lin_reg.coef_)
```
